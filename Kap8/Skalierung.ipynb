```python
import numpy as np
import pandas as pd
import scipy

import matplotlib.pyplot as plt
from matplotlib import rcParams
import seaborn as sb

import sklearn
from sklearn import preprocessing
from sklearn.preprocessing import scale
```

```python
address = 'mtcars.csv'
cars = pd.read_csv(address)
cars.columns = ['car_names','mpg','cyl','disp', 'hp', 'drat', 'wt', 'qsec', 'vs', 'am', 'gear', 'carb']
```

```python
cars[['mpg']].plot()
```

```python
cars[['mpg']].describe()
```

```python
cars[['mpg']]
```

```python
scaled = preprocessing.MinMaxScaler()
```

```python
scaled_mpg=scaled.fit_transform(cars[['mpg']])
```

```python
plt.plot(scaled_mpg)
```

```python
st_mpg=scale(cars[['mpg']], axis=0, with_mean=False, with_std=False)
```

```python
plt.plot(st_mpg)
```

```python
st_mpg=scale(cars[['mpg']], axis=0, with_mean=False, with_std=True)
```

```python
plt.plot(st_mpg)
```

```python
st_mpg=scale(cars[['mpg']], axis=0, with_mean=True, with_std=True)
```

```python
plt.plot(st_mpg)
```
